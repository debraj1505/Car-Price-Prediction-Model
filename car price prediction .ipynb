```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Data Preprocessing

```python
# outler removal for selling price
df_processed = df.drop('name', axis=1)


Q1_price = df_processed['selling_price'].quantile(0.25)
Q3_price = df_processed['selling_price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_bound_price = Q1_price - 1.5 * IQR_price
upper_bound_price = Q3_price + 1.5 * IQR_price
df_processed = df_processed[(df_processed['selling_price'] >= lower_bound_price) & (df_processed['selling_price'] <= upper_bound_price)]
```

```python
# outlier removal for km driven
Q1_km = df_processed['km_driven'].quantile(0.25)
Q3_km = df_processed['km_driven'].quantile(0.75)
IQR_km = Q3_km - Q1_km
lower_bound_km = Q1_km - 1.5 * IQR_km
upper_bound_km = Q3_km + 1.5 * IQR_km
df_processed = df_processed[(df_processed['km_driven'] >= lower_bound_km) & (df_processed['km_driven'] <= upper_bound_km)]
```

```python
print(f"Shape of data after outlier removal: {df_processed.shape}")
```

```python
X = df.drop('selling_price', axis=1)
y = df['selling_price']
```

```python
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_features = encoder.fit_transform(X[categorical_cols])
encoded_feature_names = encoder.get_feature_names_out(categorical_cols)
encoded_df = pd.DataFrame(encoded_features, columns=encoded_feature_names, index=X.index)
```

```python
X_processed = pd.concat([X[numerical_cols], encoded_df], axis=1)
```

```python
scaler = StandardScaler()
X_processed[numerical_cols] = scaler.fit_transform(X_processed[numerical_cols])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
```

```python
print(X_train.head())
```

Model Development

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, y_train)
```

```python
y_pred = linear_reg_model.predict(X_test)
print(y_pred[:10])
```

```python
print(y_test[:10].values)
```

Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2_rf= r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (RÂ²): {r2_rf:.2f}")
```
